```python
import numpy as np

t = np.arange(0., 5., 0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title('라인 플롯에서 여러 개의 선')
plt.plot(
    t, t, 'r--',
    t, t ** 2, 'bs:',
    t, t ** 3, 'g^-',
)
plt.show()
```

```python
plt.title('라인 플롯에서 여러 개의 선')
charts = [t, t, 'r--',
          t, t ** 2, 'bs:',
          t, t ** 3, 'g^-',]
plt.plot(*charts)
plt.show()
plt.title('라인 플롯에서 여러 개의 선')
chart1 = [t, t, 'r--']
chart2 = [t, t ** 2, 'bs:']
chart3 = [t, t ** 3, 'g^-']
plt.plot(*chart1, *chart2, *chart3)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title('여러 개의 plot 명령으로 한 그림에 겹쳐서 표현')
chart1 = [t, t, 'r--']
chart2 = [t, t ** 2, 'bs:']
chart3 = [t, t ** 3, 'g^-']
plt.plot(t, t, 'r--')
plt.plot(t, t ** 2, 'bs:')
plt.plot(t, t ** 3, 'g^-')
# for i in range(10):
#   plt.plot((np.arange(4) + (1 if not i % 2 else -1)) ** (i+2))
# plt.plot((np.arange(4) + 1) ** 2)
# plt.plot((np.arange(4) - 1) ** 3)
# [plt.plot((np.arange(4) + (1 if not i % 2 else -1)) ** (i+2)) for i in range(3)]
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
X = np.linspace(-np.pi, np.pi, 256)
X
```

```python
# 코사인, 사인 곡선
C, S = np.cos(X), np.sin(X)
C, S
```

```python
plt.title('삼각함수 plot')
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
plt.title('삼각함수 plot')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.legend(loc=2)  # https://kongdols-room.tistory.com/87
plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # len(x) 길이 같아야함
plt.title('바 차트')
plt.xticks(x, xlabel) # x축의 개별 값에다가 이름 붙여줌
plt.yticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기
plt.bar(x, y, color=['red', 'green', 'blue'])
plt.show()
```

```python
labels = ['호박파이', '사과파이', '정어리파이', '엄마손파이']
# sizes = [15, 30, 45, 10]
sizes = [15, 50, 72, 10]
# colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
colors = ['purple', 'gold', 'lightskyblue', 'lightcoral']
# explode = (0, 0.1, 0.2, 0.3)
explode = (0.1, 0.1, 0.1, 0.3)
plt.title("파이 차트")
# plt.pie(sizes)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=45)
plt.axis('equal')
plt.show()
```

```python
# Numpy
import numpy as np
# Pandas
import pandas as pd
# Matplotlib
import matplotlib.pyplot as plt
# Seaborn
import seaborn as sns

np.random.seed(2023)
df1 = pd.DataFrame(
    np.random.randn(100, 3), # 100 x 3 정규분포를 따르는 난수 배열
    index=pd.date_range('2023.03.15', periods=100), # 시작점과 갯수 -> n개(일)의 날짜 범위
    # 열 인덱스
    columns=list('ABC'),
).cumsum() # 행->누적합
# ).cumprod() # 행->누적곱
df1
```

```python
df1.plot()
```

```python
# https://pandas.pydata.org/pandas-docs/version/1.3/reference/api/pandas.DataFrame.plot.html
# https://wikidocs.net/159927
df1.plot() # plt.plot()
# 1. plot 안에 옵션을 넣어서 처리
# 2. plt를 불러서 처리
plt.title('df 내장 plt')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

```python
iris : pd.DataFrame = sns.load_dataset('iris') # 붓꽃
titanic : pd.DataFrame = sns.load_dataset('titanic') # 타이타닉호 데이터
```

```python
# kind = 종류 => 해당 종류 plot 교체
# iris['sepal_length'][:20].plot(kind='bar', rot=45, ylim=(4, 6), grid=True, subplots=True)
iris[:20].plot(kind='bar', rot=45, grid=True, subplots=True) # 열 별로 subplots 쪼갤거임?
# iris[:20].plot(kind='bar', rot=45, grid=True) # 열 별로 subplots 쪼갤거임?
plt.title('꽃받침의 길이 시각화')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
plt.tight_layout()
plt.show()
```

```python
# kind 를 사용할 필요 없이 plot 다음에 직접적으로 해당 차트의 메소드 호출해서 사용
iris[:5].plot.bar(rot=0)
plt.ylim(0, 7)
plt.show()
```
